```python
#The game board
board = [0,1,2,3,4,5,6,7,8]
board

def show():
    print(board[0],'|',board[1],'|',board[2])
    print('----------')
    print(board[3],'|',board[4],'|',board[5])
    print('----------')
    print(board[6],'|',board[7],'|',board[8])
    
def checkLine(char,spot1,spot2,spot3):
    if board[spot1] == char and board[spot2] == char and board[spot3] == char:
        return True
    
def checkAll(char):
    if checkLine(char,0,1,2):
        return True
    if checkLine(char,3,4,5):
        return True    
    if checkLine(char,6,7,8):
        return True
    if checkLine(char,0,3,6):
        return True
    if checkLine(char,1,4,7):
        return True
    if checkLine(char,2,5,8):
        return True
    if checkLine(char,0,4,8):
        return True
    if checkLine(char,2,4,6):
        return True  
    
y = 0    
print("Welcome to tic tac toe")
option1 = input("please enter x or o for player1")

while option1 != 'x' and option1 != 'o':
    option1 = input("Please enter a valid option")
    
if option1 == 'x':
    option2 = 'o'
else:
    option2 = 'x'
    
print("player1 takes",option1,"and player2 takes",option2)   

while True:
   
    player1 = input("select a spot for player1 ")
    player1 = int(player1)
    
    if board[player1] != 'x' and board[player1] != 'o':
        board[player1] = option1
        show()
    else:
        #print("the spot is taken")
        while y == 0:
            player1 = input("spot is taken so please select a spot for player1 ")
            player1 = int(player1)
            if board[player1] != 'x' and board[player1] != 'o':
                board[player1] = option1
                show()
                break
            else:
                y = 0
            
    
    if checkAll('x') == True:
        print("----Player1 wins------")
        break
    
    isFull = True
    for index in range(0,9):
        if board[index] != 'x' and board[index] != 'o':
            isFull = False
            break
    if isFull == True:
        print("Its a Tie")
        break
        
    player2 = input("select a spot for player2 ")
    player2 = int(player2)
    
    if board[player2] != 'x' and board[player2] != 'o':
        board[player2] = option2
        show()
    else:
        while y == 0:
            player2 = input("spot is taken so please select a spot for player2 ")
            player2 = int(player2)
            if board[player2] != 'x' and board[player2] != 'o':
                board[player2] = option2
                show() 
                break
            else:
                y = 0
           
            
    if checkAll('o') == True:
        print("----Player2 wins------")
        break
                
    isFull = True
    for index in range(0,9):
        if board[index] != 'x' and board[index] != 'o':
            isFull = False
            break
    if isFull == True:
        print("Its a Tie")
        break            
              
                   
    
            
```
